## Matplotlib ve Seaborn ile Görselleştirme

Bu notebook, turizm odaklı verileri kullanarak NumPy tabanlı görsel veri temsili ve Matplotlib / Seaborn ile temel grafik çıktılarını içerir. RGB görsel üzerinde kanal bazlı analiz yapılırken, tatilci verisiyle ısı haritası, scatter plot ve boxplot gibi tekniklerle çalışılmıştır.

1. Kütüphaneleri Yükleyelim

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

2. RGB Görselden Histogram Çıktısı

```python
rgb_gorsel = np.random.randint(0, 256, (100, 100, 3), dtype=np.uint8)

plt.hist(rgb_gorsel[:, :, 0].ravel(), bins=256, color="red", alpha=0.5, label="Kırmızı")
plt.hist(rgb_gorsel[:, :, 1].ravel(), bins=256, color="green", alpha=0.5, label="Yeşil")
plt.hist(rgb_gorsel[:, :, 2].ravel(), bins=256, color="blue", alpha=0.5, label="Mavi")
plt.legend()
plt.title("RGB Kanal Dağılımı")
plt.xlabel("Piksel Değeri")
plt.ylabel("Frekans")
plt.show()
```

3. Isı Haritası: Otel Doluluk Oranları

```python
doluluk_orani = np.random.rand(10, 12) * 100
plt.figure(figsize=(10, 6))
sns.heatmap(doluluk_orani, cmap="YlOrBr", annot=True, fmt=".1f")
plt.title("10 Otel Üzerinden 12 Günlük Doluluk Isı Haritası")
plt.xlabel("Gün")
plt.ylabel("Otel No")
plt.show()
```

4. Scatter Plot: Tatilci Yaşı vs Kalma Süresi

```python
data = pd.DataFrame({
    "Yaş": np.random.randint(18, 70, size=50),
    "Kalınan Gün": np.random.randint(1, 15, size=50)
})
sns.scatterplot(data=data, x="Yaş", y="Kalınan Gün")
plt.title("Tatilcilerin Yaş vs Kalma Süresi")
plt.show()
```

5. Boxplot: Otel Puanlarının Dağılımı

```python
puanlar = np.random.normal(loc=4.0, scale=0.8, size=100)
sns.boxplot(x=puanlar)
plt.title("Otel Puanlarının Boxplot Görünümü")
plt.xlabel("Puan")
plt.show()
```
